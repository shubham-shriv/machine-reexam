Q:-The goal of this analysis is to predict whether an individual earns more than $50K per year based on demographic and employment-related features, using a dataset of approximately 32,000 observations. The dataset contains both numerical and categorical variables, and has missing values denoted by "?". The ob	imputing missing values and encoding categorical variables, before using various classification algorithms to predict income level. This analysis aims to determine which algorithm performs the best.



```python
# python libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("adult.csv") #loading the dataset
```

```python
df[df=='?']=np.nan
```

```python
df.isnull().sum()  # finding the null values
```

```python
for col in ['workclass', 'occupation', 'native.country']:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
X = df.drop(['income'], axis=1)

y = df['income']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = pd.DataFrame(scaler.fit_transform(X_train), columns = X.columns)

X_test = pd.DataFrame(scaler.transform(X_test), columns = X.columns)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)

print('Logistic Regression accuracy score with all the features: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Accuracy 
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
